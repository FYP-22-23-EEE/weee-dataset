### Imports


```python
import os
import json
from envyaml import EnvYAML
import pandas as pd
import tsfel
import datetime
import numpy as np
```

### Accessing datasets


```python
def read_csv(path):
    return pd.read_csv(path)


def to_csv(df, path):
    df.to_csv(path, index=False)
```

E4 dataset directories


```python
VAR_ENV = EnvYAML("../../env.yaml")
path_list = [os.path.join("../..", VAR_ENV["dataset.path"], VAR_ENV["dataset.version"],
                          f"P{i:02d}", "E4") for i in range(1, VAR_ENV["dataset.participants"] + 1)]
```

### Dataframes


```python
for path in path_list:
    # dataset manipulation
    df_temp = pd.read_csv(os.path.join(path, 'ACC.csv'))
    initial_time = float(df_temp.columns[0])
    frequency = float(df_temp.iloc[0, 0])
    df_temp = df_temp.drop([0])
    df_temp.columns = ["acc_x", "acc_y", "acc_z"]

    # adding timestamps
    # timestamps = [datetime.datetime.fromtimestamp(
    #     initial_time+(i/frequency), datetime.timezone(datetime.timedelta(hours=0))) for i in range(len(df_temp))]
    # timestamps = [initial_time+(i/frequency) for i in range(len(df_temp))]
    # df_temp.insert(0, "timestamp", timestamps, True)
    df_temp["acc_x"] = df_temp["acc_x"]/64
    df_temp["acc_y"] = df_temp["acc_y"]/64
    df_temp["acc_z"] = df_temp["acc_z"]/64

    features = {
        'statistical': {
            'minimum_wristband_wrist_acc': {'complexity': 'constant',
                    'description': 'Computes the minimum value of the signal.',
                    'function': 'tsfel.calc_min',
                    'parameters': '',
                    'n_features': 1,
                    'use': 'yes'},
            'maximum_wristband_wrist_acc': {'complexity': 'constant',
                    'description': 'Computes the maximum value of the signal.',
                    'function': 'tsfel.calc_max',
                    'parameters': '',
                    'n_features': 1,
                    'use': 'yes'},
            'mean_wristband_wrist_acc': {'complexity': 'constant',
                     'description': 'Computes the mean value of the signal.',
                     'function': 'tsfel.calc_mean',
                     'parameters': '',
                     'n_features': 1,
                     'use': 'yes',
                     'tag': 'inertial'},
            'median_wristband_wrist_acc': {'complexity': 'constant',
                       'description': 'Computes median of the signal.',
                       'function': 'tsfel.calc_median',
                       'parameters': '',
                       'n_features': 1,
                       'use': 'yes'},
            'standard_deviation_wristband_wrist_acc': {'complexity': 'constant',
                                   'description': 'Computes standard deviation of the signal.',
                                   'function': 'tsfel.calc_std',
                                   'parameters': '',
                                   'n_features': 1,
                                   'use': 'yes'},
            'variance_wristband_wrist_acc': {'complexity': 'constant',
                         'description': 'Computes variance of the signal.',
                         'function': 'tsfel.calc_var',
                         'parameters': '',
                         'n_features': 1,
                         'use': 'yes'}
        }
    }

    df = tsfel.time_series_features_extractor(features, df_temp, fs=32, window_size=160, header_names=["acc_x", "acc_y", "acc_z"])

    # saving
    to_csv(df, os.path.join("/home/darshana/Projects/FYP/weee-dataset/data/v2/",
           f"P{path_list.index(path)+1}_modified_ACC.csv"))
```
